```python
import numpy as np
#how many family friendly movies have IMDB rating greater than 9.0
certu_imdb9 = np.sum((df['Certificate'] == 'U') & (df['IMDB_Rating'] >= 9.0))

#how many movies are directed by Christopher Nolan 
chrisn_movies = np.sum(df['Director'] == 'Christopher Nolan')

#how many movies have IMDB rating greater than 9.0 and Meta critic score above 90
imdb9_ms90 = np.sum((df['IMDB_Rating'] >= 9.0) & (df['Meta_score'] >= 90))

#No. of movies starring Tom Hanks
tomh_movies = np.sum((df['Star1'] == 'Tom Hanks') ^ (df['Star2'] == 'Tom Hanks') ^ (df['Star3'] == 'Tom Hanks') ^ (df['Star4'] == 'Tom Hanks'))

#No. of Drama movies
drama_movies = df['Genre'].value_counts()['Drama']
```
